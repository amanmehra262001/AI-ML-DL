```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Pandas Technical Analysis Library
import pandas_ta

# sklearn model selection
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# sklearn statistics
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

## Visualising Data

```python
df = pd.read_csv('TSLA.csv')
print(df.head())
print(df.info())
# Checking Nulls
print(df.isnull().sum())
# Statistical Data
print(df.describe())
# Checking Columns
print(df.columns)
```

```python
# Since this is regression problem and not classification problem creating a scatter graph rather than countplot makes more sense
plt.scatter(df['Date'], df['Adj Close'])
plt.ylabel('Adj Close')
plt.xlabel('Date')
plt.title('Daily Adj Close Data')
plt.show()
```

## Preprocessing Data

```python
# Reindex Data using a DatetimeIndex
df.set_index(pd.DatetimeIndex(df['Date']), inplace=True)

# Keep only 'Adj Close' value
df = df[['Adj Close']]

print(df)
```

```python
#Linear regression requires a series of assumptions to be made to be effective.
#One of these assumptions is that variables in the data are independent.
#For Time Series data this is often a problem since our observed values are longitudinal in nature—meaning they are observed values for the same thing, recorded in sequence. This produces a characteristic called autocorrelation which describes how a variable is somehow related to itself (self-related.)
#Autocorrelation analysis is useful in identifying trends like seasonality or weather patterns. When it comes to extrapolating values for price prediction, however, it is problematic. The takeaway here is that our date values aren’t suitable as our independent variable and we need to come up with something else and use the adjusted close value as the independent variable. Fortunately, there are some great options here.
#Adding Technical Indicators
```

### Adding Technical Indicators

```python
# Technical indicators are calculated values describing movements in historic pricing data for securities like stocks, bonds, and ETFs. 
# Commonly used technical indicators include moving averages (SMA, EMA, MACD), the Relative Strength Index (RSI), Bollinger Bands (BBANDS), and several others.

# Add EMA to dataframe by appending
# Note: pandas_ta integrates seamlessly into our existing DataFrame

df.ta.ema(close='Adj Close', length=10, append=True)
print(df)
```

```python
# # Drop the first n-rows that contains the NaN values, there are other methods available to substitute NaN values but for our use case nothing suits better than just dropping the NaN values
df = df.iloc[10:]
print(df)
```

```python
# Plotting EMA and Adjusted closing on single plot
plt.plot(df.index, df['Adj Close'], label='Adj Close')
plt.plot(df.index, df['EMA_10'], color='green', label = 'EMA_10')
plt.ylabel('Adj Close')
plt.xlabel('Date')
plt.title('Daily Adj Close and EMA_10 Data')
plt.legend(labels=['Adj Close', 'EMA_10'], loc='upper right')
plt.show()
```

## Model Creation

```python
# Splitting the data
X_train, X_test, Y_train, Y_test = train_test_split(df[['Adj Close']], df[['EMA_10']], test_size=0.2) 
```

```python
# Creating Model Instance
model = LinearRegression()
# Training the Model
model.fit(X_train, Y_train)
# Testing the model
Y_pred = model.predict(X_test)
```

## Metrics

```python
# Printout relevant metrics
print("Model Coefficients:", model.coef_)
print("Mean Absolute Error:", mean_absolute_error(Y_test, Y_pred))
print("Coefficient of Determination:", r2_score(Y_test, Y_pred))
```

```python
# Plotting Y_test with Y_pred line on single plot
plt.scatter(X_test, Y_test)
plt.plot(X_test, Y_pred, color='red')
plt.title('Actual v/s Predicted Data')
plt.show()
```
